```python
# Figuring out the distribution of star ratings 
vgames_review_df1.Rating.value_counts()
```

```python
vgames_review_df1['Rating'].sum()
```

**Rating-wise count of reviews**

```python
# visualizing distribution of overall star ratings
values = [120185, 54804, 28275, 14853, 13663]
labels = [5,4,3,2,1]
# Use hole to create a donut-like pie chart
fig = go.Figure(data=[go.Pie(values=values,labels=labels,hole=.3)])
fig.update_layout(
    title_text="RATINGS DISTRIBUTUOIN (Video Games)",title_x=0.5,font=dict(size=15),
    # Add annotations in the center of the donut pies.
    annotations=[dict(text='Total-947145', x=0.5, y=0.5, font_size=13, showarrow=False)])
fig.show()
```

The reviews with 5-ratings are more with 51.9%. Whereas, reviews with 1-rating are less with 5.89%.

```python
# visualizing Count of Overall Star Ratings
labels = [5,4,3,2,1]
count = []
for i in vgames_review_df1.Rating.value_counts():
    count.append(i)

fig = px.bar(x=labels, y=count, color=vgames_review_df1.Rating.unique(), text_auto=True,title='COUNT OF THE RATINGS (Video Games)')
fig.update_traces(showlegend=False,textfont_size=14)
fig.update_layout(yaxis_title='RATING COUNTS', xaxis_title='RATINGS', font=dict(size=12, color="black"), title_x=0.5)
fig.update(layout_coloraxis_showscale=False)
fig.show()
```

Approximately there are 1,20,000 reviews with 5-ratings.

**Month-wise count of reviews**

```python
# visualizing Count of monthly reviews
labels = ['January', 'February', 'March', 'April', 'May', 'June', 'July','August', 'September', 'October', 'November', 'December']
count = []
for i in vgames_review_df1.month.value_counts().sort_index():
    count.append(i)

fig = px.bar(x=labels, y=count, color=labels, text_auto=True,title='Count of Monthly Reviews (Video Games)')
fig.update_traces(showlegend=False,textfont_size=14)
fig.update_layout(yaxis_title='Number of reviews', xaxis_title='Months',font=dict(size=15, color="black"), title_x=0.5)
fig.update(layout_coloraxis_showscale=False)
fig.show()
```

The maximum number of reviews are seen in december month.

**Top-15 most reviewed products.**

```python
# visualization of most reviewed products by unique reviewers 
df1 = vgames_review_df1.groupby('asin').count().loc[:,'reviewerID'].sort_values(ascending=False).head(15)

fig = px.bar(x=df1.index, y=df1, color=df1.index, text_auto=True,title='15 Most Reviewed Products (Video Games)')
fig.update_traces(showlegend=False,textfont_size=14)
fig.update_layout(yaxis_title='Number of reviewers', xaxis_title='Product ID',font=dict(size=15, color="black"), title_x=0.5)
fig.update(layout_coloraxis_showscale=False)
print("Total number of products :",vgames_review_df1.asin.unique().shape[0])
fig.show()
```

The product ID with B00BGA9WK2 has maximum number of reviews

```python


# visualization of most reviewed products by unique reviewers 
df1 = vgames_review_df1[vgames_review_df1['Rating']==5].groupby('asin').count().loc[:,'Rating'].sort_values(ascending=False).head(10)

fig = px.bar(x=df1.index, y=df1, color=df1.index, text_auto=True,title='10 Highly Positive Rated Products Reviewed (Video Games)')
fig.update_traces(showlegend=False,textfont_size=14)
fig.update_layout(yaxis_title='Number of reviewers', xaxis_title='Product ID',font=dict(size=12, color="black"), title_x=0.5)
fig.update(layout_coloraxis_showscale=False)
print("Total number of products :",vgames_review_df1.asin.unique().shape[0])
fig.show()
```

```python

# visualization of most reviewed products by unique reviewers 
df1 = vgames_review_df1[vgames_review_df1['Rating']==1].groupby('asin').count().loc[:,'Rating'].sort_values(ascending=False).head(10)

fig = px.bar(x=df1.index, y=df1, color=df1.index, text_auto=True,title='10 Negatively Rated Products Reviewed (Video Games)')
fig.update_traces(showlegend=False,textfont_size=14)
fig.update_layout(yaxis_title='Number of reviewers', xaxis_title='Product ID',font=dict(size=12, color="black"), title_x=0.5)
fig.update(layout_coloraxis_showscale=False)
print("Total number of products :",vgames_review_df1.asin.unique().shape[0])
fig.show()
```
